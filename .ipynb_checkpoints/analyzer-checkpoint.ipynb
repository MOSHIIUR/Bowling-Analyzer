```python
import math
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
class CricketAnalyzer:
    
    def __init__(self, root, players, bowling_action, movement, keypoints):
        self.root = root
        self.file_path = ''
        self.players = players
        self.bowling_action = bowling_action
        self.movement = movement
        self.keypoints= keypoints
        self.max_frame = 0
        
    def calculate_max_frame(self):
        for file in os.listdir(self.file_path):
            if file.endswith('.csv'):
                df = pd.read_csv(os.path.join(self.file_path, file))
                self.max_frame = max(self.max_frame, df['frame'].count())
         
    def calculate_angle(self, file):
        df = pd.read_csv(file)
        data = [np.array(df[key]) for key in self.keypoints]

        AB = np.column_stack((data[2] - data[0], data[3] - data[1])) 
        AC = np.column_stack((data[2] - data[4], data[3] - data[5]))

        dot_product = np.sum(AB * AC, axis=1)
        norm = np.linalg.norm(AB, axis=1)*np.linalg.norm(AC, axis=1)

        angle_rad = np.arccos(dot_product/norm)
        angle_degrees = np.degrees(angle_rad)
        
        return angle_degrees
        
                
    def get_mean_angles(self):
        
        angles = self.get_angle()
        all_mean_angles = {}
        
        for player, player_angles in angles.items():
            max_length = max(len(angle) for angle in player_angles)
            adjusted_angles = [np.pad(angle, (max_length - angle.size, 0), 'constant', constant_values=np.nan)
                               for angle in player_angles]
            
            all_mean_angles[player] = np.nanmean(adjusted_angles, axis=0)
        
        return all_mean_angles

    
    def get_angle(self):
        all_angles = {}
        for player in self.players:
            self.file_path = os.path.join(self.root, player, 'csvs', self.bowling_action)
            self.calculate_max_frame()
                      
            angles = []
            for file in os.listdir(self.file_path):
                if file.endswith('.csv'):
                    file_name = os.path.join(self.file_path, file)
                    angle = self.calculate_angle(file_name)
                    angles.append(angle)
            all_angles[player] = angles

            self.max_frame = 0
        return all_angles
    
    def plot_angle(self, angles):
        for player, player_angles in angles.items():
            max_length = max(len(angle) for angle in player_angles)
                
            for i, angle in enumerate(player_angles):
                print(f'{player} {self.bowling_action} {self.movement}')
                print(angle)
                frame_diff = max_length - angle.size
                xs = np.arange(angle.size) + frame_diff
#                 plt.plot(xs, angle, label=f'{self.bowling_action}-{i}')
                plt.plot(angle, label=f'{self.bowling_action}-{i}')
                plt.title(f'{player} {self.bowling_action} {self.movement}')
                plt.xlabel('Frames(Adjusted)')
                plt.ylabel('Angles(Degree)')
                plt.legend()
    #             plt.xlim(max_length/2, max_length)
                plt.show()
                plt.close()
            
    def plot_mean_angles(self, angles):
        max_length = max(angle.size for angle in angles.values())
            
        for player, player_angle in angles.items():
            frame_diff = max_length - player_angle.size
            xs = np.arange(player_angle.size) + frame_diff
            plt.plot(xs, player_angle, label=f'{player}')
            
        plt.title(f'{self.bowling_action} Mean {self.movement}')
        plt.xlabel('Frames(Adjusted)')
        plt.ylabel('Angles(Degree)')
        plt.legend()
#         plt.xlim(max_length/2, max_length)
        plt.show()
        plt.close()

                    
                
```

```python
root_folder = r'G:\DL_enigma\Bowling-Analyzer\Data\csvs\Front'
# players = ['Jasprit_Bumrah', 'Jofra_Archer', 'Mark_Wood', 'Mitchell_Starc', 'Shaun_Tait']
players = ['Mark Wood']

bowling_action = 'SideOn' #[SideOn][Rear][Front]
movement = 'Right Knee Angle' # Right/left (leg/hand), Trunk Flexion, etc.
keypoints = ['x_24', 'y_24', 'x_26', 'y_26', 'x_28', 'y_28']

player = CricketAnalyzer(root_folder, players, bowling_action, movement, keypoints)
```

```python
# get_all_angle
angles = player.get_angle()
player.plot_angle(angles)
```

```python
# get_mean
mean_angles = player.get_mean_angles()
player.plot_mean_angles(mean_angles)
```
